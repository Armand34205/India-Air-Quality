**The objective of this notebook is to explore the air quality of India's major cities, and develop a story based off the data. The priority is to build a narrative about the data based on EDA findings. The dataset being worked with contains Air Quality data from all around India. A key factor will be to analyze AQI (Air Quality Index) levels throughout India. AQI is used by governments to communicate to the public how polluted the air currently is. AQI is measured on a scale of 0 to 500, with the higher number meaning more pollution.**

*Questions to Consider: Which are the most polluted cities in India? What levels of pollution are seen in these cities? Which particle matters seem most impactful when determining air quality?*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('city_day.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
def missing_values(df):

    missing_val = df.isnull().sum()
    
    missing_val_pct = 100 * df.isnull().sum() / len(df)
    
    table = pd.concat([missing_val, missing_val_pct], axis = 1)
    
    missing_values_renamed = table.rename(columns = {0: '# of Missing Values', 1 : '% of Total Values'})
    
    missing_values_renamed = missing_values_renamed[missing_values_renamed.iloc[:,1] != 0].sort_values(by = '% of Total Values', ascending = False)
    
    return missing_values_renamed
```

```python
missing_values(df)
```

```python
df['AQI'].fillna(method = 'bfill',inplace = True)
df['AQI_Bucket'].fillna(method = 'bfill',inplace = True)
```

```python
AQI_avg = df.groupby(['City']).AQI.agg(['mean','max','min']).reset_index()
sorted_1 = AQI_avg.sort_values('mean',ascending=False)
sorted_1
```

```python
AQI_Bucket_max = df.groupby(['City']).AQI_Bucket.agg(['max']).reset_index()
AQI_Bucket_max.sort_values('max',ascending=False)
```

```python
plt.figure(figsize=(35,15))
plt.bar(AQI_avg['City'],AQI_avg['mean'],data='sorted_1')
plt.xlabel('City',size=20)
plt.ylabel('Average AQI',size=20)
plt.title('Average AQI in India by City',size=25)
plt.show()
```

```python
df['PM10'].fillna(method = 'bfill',inplace = True)
```

```python
PM10_data = df.groupby(['City']).PM10.agg(['mean']).reset_index()
PM10_data.sort_values(by = 'mean', ascending=False)
```

**Analysis / Conclusions:**

Ahmedabad has the highest average AQI, meaning it has the highest level of pollution on any given day. With an average AQI over 300, Ahmedabad sits in the 'hazardous' category of the AQI scale. This is problematic to all people living there, and will generally affect day to day behavior within the city. Exposure to outdoor air and activties will most likely be limited, and will also promote mask wearing and other hygenic behaviors.

However, Ahmedabad is not the only city with pollution issues. The majority of big cities in India that are captured in this data have had days classified as 'Very Poor' in the AQI Bucket. That means that they have experienced days with AQI levels consistent with Very Poor quality. This raises questions about the cause of pollution in India's largest cities. Further research and investigation may be able to shed light on the largest causes of this poor air quality. Common factors would most likely include an increasing population, vehicle emmissions, and industrial activities. It is safe to say that the largest cities in India have a serious problem with air pollution, and must make changes in order to improve quality of life there.

Another piece of data to consider is the volume of PM10 particles in the air, as shown in the table above. PM10 is a measure of how many particles that are 10 microns or less in size, are in the air. PM does not include gas pollutants like ozone and NO2. PM directly translates to particulate matter, so things like dust, airborne bacteria, construction dust and coal particles are captured in this data. The WHO estimates that PM2.5 and PM10 affect mopre people than any other pollutant, and thus serve as an effective metric when analyzing the quality of life in an area.

As evident from the table above, levels of PM10 flucuate greatly throughout cities in India. Cities like Delhi and Ahmedabad have both high levels of PM10 and high AQI scores. This gives sufficient evidence to say that they are the most polluted cities within India, and pose the greatest risk to their citizens. Areas with high levels of PM10 can be difficult for people with asthma or other respiratory diseases. Life in these citizens may prove difficult for even healthy individuals.

In conculsion, the Indian government and other world health organizations need to make it a priority to clean up the pollution within India. Activities like burning coal, wood, and crop waste can cause high levels of PM10 and need to be reevaluated. A rapidly growing population certainly does not help matters, as there in an industrial need for construction and energy. However, a revamped approach to dealing with air pollution is necessary given the health risks associated with PM10. India will continue to develop as a nation, but would benefit greatly from a more environmentaly friendly approach to developing infrastructure.
